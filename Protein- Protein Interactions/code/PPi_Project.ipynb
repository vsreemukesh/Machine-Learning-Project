# Filling the Null Values

```python
drug_data.fillna({'clustering.coefficient':0},inplace=True)
```

```python
data=drug_data.iloc[:,1:14] #creating object for the features in data
```

```python
data.head()
```

# Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
feature_scaler=StandardScaler()
data1=feature_scaler.fit_transform(data)
```

```python
data1.tolist() #converting to list
```

```python
data1=DataFrame.from_records(data1) #converting to Dataframe
```

```python
type(data1) #checking type of data
```

```python
proteins=nd.DataFrame(drug_data.iloc[:,0]) #Assigning nodes   to proteins object
```

```python
classes=nd.DataFrame(drug_data.iloc[:,14]) #Assigning labels  to classes object
```

```python
f=[proteins,data1,classes]
```

```python
data_scaled=nd.concat(f,axis=1) #scaling the data 
```

```python
data_scaled.head()
```

```python
data_scaled.columns=[drug_data.columns]
```

```python
data_scaled.head()
```

# Stratified Sampling

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split= StratifiedShuffleSplit(n_splits=1,test_size=.30,random_state=0)
```

```python
for train_index, test_index in split.split(data_scaled,data_scaled['classes']):
    strat_train_set = data_scaled.loc[train_index]
    strat_test_set = data_scaled.loc[test_index]
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=500,random_state=47,max_depth=9)
model1.fit(strat_train_set.iloc[:,1:14],strat_train_set['classes'])
```

```python
predict1 = model1.predict(strat_test_set.iloc[:,1:14])
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(strat_test_set['classes'],predict1) #checking accuracy
```

```python
nd.DataFrame(confusion_matrix(strat_test_set['classes'],predict1))
```

# Grid Search on Random Forest

```python
import time
from sklearn.model_selection import GridSearchCV
parameters_grid={'n_estimators':[150,200,250,300,500],'max_depth':[5,8,10,20],'bootstrap':[True,False]}
grid=GridSearchCV(estimator=model1,param_grid=parameters_grid,cv=3,n_jobs=-1)
starttime=time.time()
grid_result_rf=grid.fit(strat_train_set.iloc[:,1:14],strat_train_set['classes'])
print(grid_result_rf.best_score_,grid_result_rf.best_params_)
time.time()-starttime
```

```python
predict_grid_rf=grid_result_rf.predict(strat_test_set.iloc[:,1:14])
accuracy_score(strat_test_set['classes'],predict_grid_rf)
```

```python
nd.DataFrame(confusion_matrix(strat_test_set['classes'],predict_grid_rf))
```

# Stacking using RF,GB,XGBOOST algorithms

```python
from mlxtend.classifier import StackingClassifier,StackingCVClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
from sklearn.model_selection import GridSearchCV
```

```python
lr = LogisticRegression()
main_model = StackingClassifier(classifiers=[model1,model2,model7], meta_classifier=lr)
```

```python
for model, label in zip([model1,model2,model7,main_model], 
                      [  'Random Forest',
                         'Gradient Bosting',
                         'Xgboost',
                         'StackingClassifier']):
    scores = model_selection.cross_val_score(model, strat_test_set.iloc[:,1:14],strat_test_set['classes'], cv=3, scoring='accuracy')
    print("Accuracy: %0.2f (+/- %0.2f) [%s]" 
          % (scores.mean(), scores.std(), label))
```
